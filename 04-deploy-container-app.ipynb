# 04 — Deploy an App from a Container Image

The easiest way to deploy an application on OpenShift is to point it at a pre-built container image.
We will deploy a simple welcome app from [quay.io](https://quay.io) and give it a public URL.

Run each cell from top to bottom by pressing **Shift + Enter**.

## Step 1 — Deploy the container image

```python
!oc new-app quay.io/eformat/welcome:latest
```

## Step 2 — Create a public route (HTTPS URL)

```python
!oc create route edge --service=welcome --port=8080 --insecure-policy=Redirect
```

## Step 3 — Get your app URL

```python
host = !oc get route welcome -o jsonpath='{.spec.host}'
print("Your app URL: https://" + host[0])
```

## Step 4 — Check the pod is running

It may take 30–60 seconds for the pod to start. Run this cell again if the status is not yet `Running`.

```python
!oc get pods -l app=welcome
```

Open the URL from Step 3 in your browser once the pod shows `Running`.
